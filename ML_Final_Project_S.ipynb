```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
```

```python
df = pd.read_csv("/content/balanced_loan_default_dataset.csv")
df
```

```python
df=df.drop(columns=['Applicant_ID','Occupation'])
df
```

```python
df=df.drop(columns=['Residence_City','Residence_State'])
df
```

```python
df_encoded = pd.get_dummies(df, drop_first=True).astype(int)
df_encoded
```

```python
X_final= df_encoded.drop("Loan_Default_Risk", axis=1)
y = df_encoded["Loan_Default_Risk"]
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.1, random_state=42)
```

```python
model_columns = list(X_final.columns)
```

```python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_final_train, y_train)
y_pred_lr = lr_model.predict(X_final_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
y_pred = lr_model.predict(X_final_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
dt_model = DecisionTreeClassifier()
dt_model.fit(X_final_train, y_train)
y_pred_dt = dt_model.predict(X_final_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
rf_model = RandomForestClassifier(class_weight='balanced')
rf_model.fit(X_final_train, y_train)
y_pred_rf = rf_model.predict(X_final_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
from xgboost import XGBClassifier
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_final_train,y_train)

y_pred = xgb_model.predict(X_final_test)

print("XGBoost Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
import joblib

# Load model and column names
model = joblib.load("rf_model.pkl")
model_columns = joblib.load("model_columns_rf.pkl")

# Indian number formatter (manual fallback)
def format_in_indian_style(n):
    s = str(int(n))
    r = ""
    if len(s) > 3:
        r = "," + s[-3:]
        s = s[:-3]
        while len(s) > 2:
            r = "," + s[-2:] + r
            s = s[:-2]
        r = s + r
    else:
        r = s
    return r

# Prediction function
def predict(loan_amt, income, age, exp, emp_yrs, res_yrs, is_single, owns_home):
    try:
        # Clean comma inputs
        loan_amt = int(str(loan_amt).replace(",", "").strip())
        income = int(str(income).replace(",", "").strip())

        data = {
            "Requested_Loan_Amount": loan_amt,
            "Annual_Income": income,
            "Applicant_Age": age,
            "Work_Experience": exp,
            "Years_in_Current_Employment": emp_yrs,
            "Years_in_Current_Residence": res_yrs,
            "Marital_Status_single": int(is_single == "Yes"),
            "House_Ownership_owned": int(owns_home == "Yes")
        }

        df = pd.DataFrame([data])
        for col in model_columns:
            if col not in df.columns:
                df[col] = 0
        df = df[model_columns]

        prob = model.predict_proba(df)[0][1] * 100
        f_income = format_in_indian_style(income)
        f_loan = format_in_indian_style(loan_amt)

        if prob >= 50:
            msg = f"""🔴 High Risk of Loan Default ({prob:.2f}%)
💰 Loan: ₹{f_loan}
💵 Income: ₹{f_income}"""
        else:
            msg = f"""🟢 Low Risk of Loan Default ({prob:.2f}%)
💰 Loan: ₹{f_loan}
💵 Income: ₹{f_income}"""

        # Pie chart
        fig, ax = plt.subplots()
        parts = [loan_amt / 1e6, income / 1e6, emp_yrs, res_yrs]
        labels = ["Loan", "Income", "Employment", "Residence"]
        colors = ['#E74C3C', '#2ECC71', '#3498DB', '#F1C40F']
        ax.pie(parts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
        ax.axis('equal')
        plt.title("📊 Contribution Breakdown")

        return msg, fig

    except Exception as e:
        return f"❌ Error: {str(e)}", None

# UI
gr.Interface(
    fn=predict,
    inputs=[
        gr.Textbox(label="💰 Requested Loan Amount (₹)", placeholder="e.g. 5,00,000"),
        gr.Textbox(label="💵 Annual Income (₹)", placeholder="e.g. 12,00,000"),
        gr.Slider(18, 70, label="🎂 Applicant Age"),
        gr.Slider(0, 40, label="🧰 Work Experience"),
        gr.Slider(0, 30, label="👨‍💼 Years in Current Employment"),
        gr.Slider(0, 30, label="🏠 Years in Current Residence"),
        gr.Radio(["No", "Yes"], label="❤️ Marital Status (Single)"),
        gr.Radio(["No", "Yes"], label="🏡 Home Ownership (Owned)")
    ],
    outputs=[
        gr.Textbox(label="📋 Risk Level"),
        gr.Plot(label="📈 Factor Chart")
    ],
    title="🏦 Loan Default Risk Detector (High/Low Only)",
    description="Only shows 🔴 High Risk or 🟢 Low Risk predictions."
).launch()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
import joblib

# Load model and column structure
model = joblib.load("lr_model.pkl")
model_columns = joblib.load("model_columns.pkl")

# Feature labels with emojis
FEATURE_LABELS = {
    "Requested_Loan_Amount": "💰 Requested Loan Amount (₹)",
    "Annual_Income": "💵 Annual Income (₹)",
    "Applicant_Age": "🎂 Applicant Age",
    "Work_Experience": "🧰 Work Experience (years)",
    "Years_in_Current_Employment": "👨‍💼 Years in Current Employment",
    "Years_in_Current_Residence": "🏠 Years in Current Residence",
    "Marital_Status_single": "❤️ Marital Status (Single)",
    "House_Ownership_owned": "🏡 Home Ownership (Owned)"
}

# Prediction function with pie chart explanation
def predict_and_explain(loan_amount, income, age, experience, job_years, residence_years, is_single, owns_home):
    input_dict = {
        "Requested_Loan_Amount": loan_amount,
        "Annual_Income": income,
        "Applicant_Age": age,
        "Work_Experience": experience,
        "Years_in_Current_Employment": job_years,
        "Years_in_Current_Residence": residence_years,
        "Marital_Status_single": int(is_single == "Yes"),
        "House_Ownership_owned": int(owns_home == "Yes")
    }

    df = pd.DataFrame([input_dict])
    for col in model_columns:
        if col not in df.columns:
            df[col] = 0
    df = df[model_columns]

    try:
        probability = model.predict_proba(df)[0][1] * 100

        # Format numbers with commas
        formatted_loan = f"{loan_amount:,.0f}"
        formatted_income = f"{income:,.0f}"

        # 🧠 Risk Interpretation + Number Display
        if probability >= 70:
            label = f"""🔴 High Risk of Loan Default ({probability:.2f}%)
💰 Loan Amount: ₹{formatted_loan}
💵 Income: ₹{formatted_income}"""
        elif 40 <= probability < 70:
            label = f"""🟠 Medium Risk of Loan Default ({probability:.2f}%)
💰 Loan Amount: ₹{formatted_loan}
💵 Income: ₹{formatted_income}"""
        else:
            label = f"""🟢 Low Risk of Loan Default ({probability:.2f}%)
💰 Loan Amount: ₹{formatted_loan}
💵 Income: ₹{formatted_income}"""

        # 📈 Explanation Pie Chart
        fig, ax = plt.subplots()
        values = [loan_amount / 1e6, income / 1e6, job_years, residence_years]
        labels = ["Loan Amount", "Income", "Job Stability", "Residence Stability"]
        colors = ['#FF5722', '#4CAF50', '#2196F3', '#FFC107']
        ax.pie(values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
        ax.axis('equal')
        plt.title("📊 Risk Factor Contribution")

        return label, fig

    except Exception as e:
        return f"❌ Prediction Error: {str(e)}", None

# Gradio UI
gr.Interface(
    fn=predict_and_explain,
    inputs=[
        gr.Number(label=FEATURE_LABELS["Requested_Loan_Amount"], value=100000),
        gr.Number(label=FEATURE_LABELS["Annual_Income"], value=300000),
        gr.Slider(18, 70, label=FEATURE_LABELS["Applicant_Age"]),
        gr.Slider(0, 40, label=FEATURE_LABELS["Work_Experience"]),
        gr.Slider(0, 30, label=FEATURE_LABELS["Years_in_Current_Employment"]),
        gr.Slider(0, 30, label=FEATURE_LABELS["Years_in_Current_Residence"]),
        gr.Radio(["No", "Yes"], label=FEATURE_LABELS["Marital_Status_single"]),
        gr.Radio(["No", "Yes"], label=FEATURE_LABELS["House_Ownership_owned"])
    ],
    outputs=[
        gr.Textbox(label="📋 Risk Assessment"),
        gr.Plot(label="📈 Contribution Chart")
    ],
    title="🏦 Loan Default Risk Predictor",
    description="Get AI-based risk level with a pie chart showing key contributing factors."
).launch()
```
